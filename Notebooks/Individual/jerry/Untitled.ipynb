```python
import pandas as pd
import numpy as np

df = pd.read_csv("../../../archive/high_diamond_ranked_10min.csv")
pd.set_option("display.max_columns", None)

df.head()
```

```python
df["blueVisionScore"] = df["blueWardsPlaced"]
df["redVisionScore"] = df["redWardsPlaced"]
df = df.drop(["blueWardsPlaced", "redWardsPlaced"], axis = 1)
```

```python
df = df.drop(["redGoldDiff", "redExperienceDiff", "redFirstBlood", "gameId", "blueTotalGold",
             "blueEliteMonsters", "blueTotalMinionsKilled", "blueTotalExperience", "redEliteMonsters",
             "redTotalGold", "redTotalExperience", "redTotalMinionsKilled", "blueDeaths", "redDeaths"], axis = 1)
df.head()
```

```python
corr = df.corr().stack().reset_index()
corr["pairs"] = corr["level_0"] + " & " + corr["level_1"]
corr = corr.loc[corr["level_0"] != corr["level_1"]]
corr = corr.drop(["level_0", "level_1"], axis = 1).set_index("pairs")
```

```python
from sklearn.model_selection import train_test_split as tts
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix

X = df.drop("blueWins", axis = 1) 
y = df["blueWins"]

X_train, X_test, y_train, y_test = tts(X, y, random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
X_train_scaled = ss.fit_transform(X_train)
X_test_scaled = ss.transform(X_test)
```

```python
tree = DecisionTreeClassifier()
tree.fit(X_train_scaled, y_train)

y_hat_train = tree.predict(X_train_scaled)
y_hat_test = tree.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score, f1_score

print("train accuracy:", accuracy_score())
```
