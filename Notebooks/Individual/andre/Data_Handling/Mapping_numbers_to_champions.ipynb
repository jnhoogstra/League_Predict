```python
import requests
from bs4 import BeautifulSoup
import json
import pandas as pd
```

```python
request = requests.get('http://ddragon.leagueoflegends.com/cdn/11.14.1/data/en_US/champion.json')
```

```python
soup = BeautifulSoup(request.text, 'lxml')
```

```python
dict_string = soup.find_all('p')[0].text
```

```python
dictionary = json.loads(dict_string)
```

```python
champions = dictionary['data']
```

```python
champions['Aatrox']
```

```python
champ_to_num = {data['name'] : int(data['key']) for data in champions.values()}
```

```python
champ_to_num
```

```python
num_to_champ = dict(zip(champ_to_num.values(), champ_to_num.keys()))
```

```python
num_to_champ
```

```python
df = pd.read_csv('../../../../archive/with_roles.csv')
```

```python
df['blueChamps'] = df['blueChamps'].map(lambda x: [int(y) for y in x.strip('][').split(', ')])
df['blueChamps'] = df['blueChamps'].map(lambda x: [num_to_champ[num] for num in x])
```

```python
df['redChamps'] = df['redChamps'].map(lambda x: [int(y) for y in x.strip('][').split(', ')])
df['redChamps'] = df['redChamps'].map(lambda x: [num_to_champ[num] for num in x])
```

```python
df.to_csv(path_or_buf = '../../../../archive/With_champ_names.csv', index = False)
```
