```python
import pandas as pd
```

```python
df1 = pd.read_csv('../../../../archive/3532_items-levels.csv')
```

```python
df2 = pd.read_csv('../../../../archive/remaining_items-levels.csv')
```

```python
combined = pd.concat([df1, df2])
```

```python
combined.to_csv('../../../../archive/items-levels.csv')
```

```python
levels = combined['levels']
```

```python
levels
```

```python
def to_dict(string):
    to_return = {}
    pairs = string.strip('}{').split(', ')
    for pair in pairs:
        pair = pair.split(': ')
        to_return[pair[0]] = int(pair[1])
    return to_return
```

```python
levels = levels.map(to_dict)
```

```python
levels
```

```python
def blues(level_dict):
    blue = []
    for x in range(1,6):
        blue.append(level_dict[str(x)])
    return blue
```

```python
blueLevels = levels.map(blues)
```

```python
def reds(level_dict):
    red = []
    for x in range(6,11):
        red.append(level_dict[str(x)])
    return red
```

```python
redLevels = levels.map(reds)
```

```python
blueLevels
```

```python
redLevels
```

```python
level_df = pd.DataFrame({'levels':levels, 'blueLevels':blueLevels,
                        'redLevels':redLevels})
```

```python
level_df
```

```python
import numpy as np
```

```python
level_df['blueMaxLevel'] = level_df['blueLevels'].map(max)
level_df['blueMinLevel'] = level_df['blueLevels'].map(min)
level_df['blueSTDLevels'] = level_df['blueLevels'].map(np.std)

level_df['redMaxLevel'] = level_df['redLevels'].map(max)
level_df['redMinLevel'] = level_df['redLevels'].map(min)
level_df['redSTDLevels'] = level_df['redLevels'].map(np.std)
```

```python
big_df = pd.read_csv('../../../../archive/With_champ_names.csv')
```

```python
all_df = pd.concat([big_df.reset_index(drop = True), level_df.reset_index(drop = True)], axis = 1)
```

```python
all_df
```

```python
all_df.to_csv(path_or_buf = '../../../../archive/All_high_diamond.csv', index = False)
```
