```python
import pandas as pd
import pickle
import numpy as np
```

```python
with open('../../../../Pickles/champ_winrates.pkl', 'rb') as f:
    champ_winrates = pickle.load(f)
    
with open('../../../../Pickles/champ_pickrates.pkl', 'rb') as f:
    champ_pickrates = pickle.load(f)
```

```python
len(champ_winrates) == len(champ_pickrates)
```

```python
winrate_df = pd.DataFrame(champ_winrates, index = ['winrates']).T
```

```python
pickrate_df = pd.DataFrame(champ_pickrates, index = ['pickrates']).T
```

```python
winrate_df.sort_index(inplace = True)
winrate_df
```

```python
pickrate_df.sort_index(inplace = True)
pickrate_df
```

```python
rates_df = pd.concat([winrate_df, pickrate_df], axis = 1)
```

```python
rates_df
```

```python
rates_df.to_csv(path_or_buf = '../../../../archive/rates_df.csv', index = False)
```

```python
rates_df.sort_values('winrates', ascending = False)
```

```python
rates_df.sort_values('pickrates', ascending = False)
```

```python
rates_df.loc[rates_df['winrates'] == rates_df['winrates'].min()]
```

```python
relevants_df = rates_df.loc[rates_df['pickrates'] >= 0.08]
```

```python
relevants_df.sort_values('winrates', ascending = False)
```

```python
rates_df['impact'] = np.sqrt(abs(0.5 - rates_df['winrates'])*rates_df['pickrates'])
```

```python
rates_df.sort_values('impact', ascending = False)
```

```python
most_impact = [62,111,145,39,517]
```

```python
df = pd.read_csv('../../../../archive/with_rates.csv')
```

```python
df.drop('champions', axis = 1, inplace = True)
```

```python
df['blueChamps'] = df['blueChamps'].map(lambda x: [int(y) for y in x.strip('][').split(', ')])
```

```python
df['redChamps'] = df['redChamps'].map(lambda x: [int(y) for y in x.strip('][').split(', ')])
```

```python
for champ in most_impact:
    df['blueHas'+str(champ)] = df['blueChamps'].map(lambda x: champ in x)
    df['redHas'+str(champ)] = df['redChamps'].map(lambda x: champ in x)
```

```python
df.to_csv(path_or_buf = '../../../../archive/with_rates_and_spec_gods.csv', index = False)
```
