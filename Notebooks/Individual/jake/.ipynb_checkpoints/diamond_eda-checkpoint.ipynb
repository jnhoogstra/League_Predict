```python
import pandas as pd
import numpy as np

from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
from sklearn.compose import ColumnTransformer
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, roc_auc_score, plot_confusion_matrix


pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("raw_data/high_diamond_ranked_10min.csv")

df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
drop_columns = ["gameId", "blueDeaths", "redDeaths", "blueEliteMonsters", "blueTotalExperience",
               "redTotalExperience", "redEliteMonsters"]

df = df.drop(columns=drop_columns, axis = 1)
```

```python
y = df["blueWins"]
X = df.drop("blueWins", axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=57)

boost_model = XGBClassifier(random_state=57, objective="reg:squarederror")

param_grid = {
    'learning_rate': [0.1, 0.2],
    'max_depth': [4, 5],
    'min_child_weight': [1, 2, 3],
    'subsample': [0.6, 0.7],
    'n_estimators': [50, 100]
}

gridsearch = GridSearchCV(boost_model, param_grid, scoring="accuracy", n_jobs=1)
gridsearch.fit(X_train, y_train)

best_parameters = gridsearch.best_params_

print("Best Parameters: ")
print(best_parameters)
```

```python
train_pred = gridsearch.predict(X_train)
test_pred = gridsearch.predict(X_test)

train_acc = accuracy_score(y_train, train_pred)
train_f1 = f1_score(y_train, train_pred)

print("Training Scores")
print("Accuracy: ", train_acc)
print("F1: ", train_f1)

test_acc = accuracy_score(y_test, test_pred)
test_f1 = f1_score(y_test, test_pred)

print("Test Scores")
print("Accuracy: ", test_acc)
print("F1: ", test_f1)
```
